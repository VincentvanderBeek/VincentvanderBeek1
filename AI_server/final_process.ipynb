```python
import pandas as pd
import sys
```

```python
df = pd.read_csv("training_dataset.csv")
```

```python
df.value_counts("emotion")
```

```python
# Check current counts for reference
print(df['emotion'].value_counts())
```

```python
# Separate disgust rows and others
df_disgust = df[df['emotion'] == 'disgust']
df_other = df[df['emotion'] != 'disgust']

# Downsample disgust to 5000 if more than 5000 rows exist
if len(df_disgust) > 5000:
    df_disgust_downsampled = df_disgust.sample(n=5000, random_state=42)
else:
    df_disgust_downsampled = df_disgust  # Keep all if less than 5000

# Combine back
df_balanced = pd.concat([df_disgust_downsampled, df_other], ignore_index=True)

# Check counts after
print(df_balanced['emotion'].value_counts())
```

```python
# Save the final processed DataFrame
df_balanced.to_csv('training_dataset.csv', index=False)
```
